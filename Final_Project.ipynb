```python
import pandas as pd

# Importing the Climbing data for the year 2014 and 2015
climbing_df = pd.read_csv("climbing_statistics.csv")

# Importing the Weather data for the year 2014 and 2015
weather_df = pd.read_csv("Rainier_Weather.csv")
```

```python
# Merging tables using the inner join with date as the key
joined_df = pd.merge(climbing_df, weather_df, on='Date', how='inner')
joined_df
```

```python
# Describing the whole joined climbers and weather data
joined_df.describe()
```

```python
# Using groupby to calculate the average temprature during a successful summit
joined_df.groupby('Succeeded',as_index=False).agg({'Temperature AVG':['mean']})
```

```python
# Using groupby to calculate the median relative humidity during a successful summit
joined_df.groupby('Succeeded',as_index=False).agg({'Relative Humidity AVG':['median']})
```

```python
# Using groupby to calculate the average wind speed during a successful summit
joined_df.groupby('Succeeded',as_index=False).agg({'Wind Speed Daily AVG':['mean']})
```

```python
# Using groupby to calculate the sum of the successful summits out of all the attempts
joined_df.groupby('Attempted',as_index=False).agg({'Succeeded':['sum']})
```

```python
# Pivot table
pivoted_df=joined_df.pivot_table(columns='Date',index='Route',values='Succeeded', aggfunc='count')
pivoted_df
```

```python
# Melting the pivot table
melted_df = joined_df.melt(id_vars='Date',value_vars=['Route', 'Succeeded'])
melted_df
```
