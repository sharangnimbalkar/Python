```python
import pandas as pd

# Importing the Climbing data for the year 2014 and 2015
climbing_df = pd.read_csv("climbing_statistics.csv")

# Importing the Weather data for the year 2014 and 2015
weather_df = pd.read_csv("Rainier_Weather.csv")
```

```python
# Merging tables using the inner join with date as the key
joined_df = pd.merge(climbing_df, weather_df, on='Date', how='inner')
joined_df
```

```python
# Using groupby to calculate the average temprature during a successful summit
joined_df.groupby('Succeeded',as_index=False).agg({'Temperature AVG':['mean']})
```

```python
# Using groupby to calculate the median relative humidity during a successful summit
joined_df.groupby('Succeeded',as_index=False).agg({'Relative Humidity AVG':['median']})
```

```python
# Using groupby to calculate the average wind speed during a successful summit
joined_df.groupby('Succeeded',as_index=False).agg({'Wind Speed Daily AVG':['mean']})
```

```python
# Using groupby to calculate the sum of the successful summits out of all the attempts
joined_df.groupby('Attempted',as_index=False).agg({'Succeeded':['sum']})
```

```python
# As there are number of dates when the climb attempts we made, we group them into months

import calendar

joined_df["Date"] = pd.to_datetime(joined_df["Date"])
joined_df["Month"] = 0
joined_df["Month"] = joined_df["Date"].dt.month

#Using the lambda function
joined_df.Month = joined_df.Month.apply(lambda x: calendar.month_abbr[x])

joined_df.head()

# An extra column of month is appended in the dataset.
```

```python
# Computing the impact of temperature on the success rate
f, ax = plt.subplots(1, 1, figsize=(15, 7))

joined_df["Success Percentage"] = joined_df["Success Percentage"] * 100

sns.lineplot(x="Month", y="Temperature AVG", data=joined_df, ax=ax)
sns.lineplot(x=joined_df["Month"], y="Success Percentage", data=joined_df, ax=ax)

ax.set_ylabel("")
f.legend(labels=["Success Rate", "Temperature"])
plt.show()

# Observation:
# 1. We observe that everytime when temperature is more the success rate is increased and vica versa
# 2. In colder climate(December, November) the success rate should be low as the mountains are harder to climb
# 3. Also, In the summer(June, July) the success rate should be high as the mountains are easier to climb
```

```python
# Further computing the line plot for success rate vs solare radiation for each month
f, ax= plt.subplots(1, 1, figsize=(15, 7))
sns.lineplot(x="Month", y="Solare Radiation AVG", data=joined_df, ax=ax)
sns.lineplot(x="Month", y="Success Percentage", data=joined_df, ax=ax)
f.legend(labels=["Solare Radiation", "Success Rate"])
ax.set_label("")
plt.legend()
plt.show()

# The success rate is more when the solar radiation is more, confirming the previous corelation between solar radiation and the temperature.
```

```python
# Further computing the line plot for success rate vs Wind speed for each month
fig, ax = plt.subplots(1, 1, figsize=(15, 7))

sns.lineplot(x="Month", y="Wind Speed Daily AVG", data=joined_df, ax=ax)
sns.lineplot(x="Month", y="Success Percentage", data=joined_df, ax=ax)
ax.set_ylabel("")
fig.legend(labels=["Wind Speed", "Success Rate"])
plt.legend()
plt.show()

# We observe that when the wind speed dips the success rate is increased and vice versa meaning that in low winds it is earier to climb.
# This confirms their negative corelation.
```

```python
# For ease in the computation, coverting the success percentage and dropping continuous variables.
y = joined_df["Success Percentage"] / 100
joined_df.drop(["Class", "Date", "Success Percentage", "Route", "Month"], axis=1, inplace=True)
X = joined_df
```

```python
# Splitting the data in train (70%) and test (30%)
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,train_size=0.7,
test_size=0.3,random_state = 12)
```

```python
# Now forming the regression models to find which model gives the best results

from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import numpy as np

#Creating classes for each model

# 1. Linear Regression Model
Linear_Reg = LinearRegression()

# 2. Ridge Regression Model
Ridge = Ridge()

# 3. Lasso Regression Model
Lasso = Lasso()

# 4. Elastic Net Regression Model
Elastic = ElasticNet()

# 5. K Nearest Neighbors Model
KNN = KNeighborsRegressor()

# 6. Decision Tree
Decision_tree = DecisionTreeRegressor(random_state = 12)

# 7. Random Forest Model
Random_forest = RandomForestRegressor(random_state= 12)
```

```python
# For ease in comparision, we will run a loop with all the models
# Computing the MSE, R2 and prediction level of each method

for model, name in zip([Linear_Reg,Lasso,Ridge,Elastic,Decision_tree,Random_forest,KNN], 
     ['Linear Regression','Ridge Regression','Lasso','Elastic Net','Decision Tree Regressor','Random Forest Regressor','KNN']):
    model1 = model.fit(Xtrain,ytrain)
    Ypredict=model1.predict(Xtest)
    print(name)
    print('Mean Absolute Error:', mean_absolute_error(ytest, Ypredict))  
    print('Mean Squared Error:', mean_squared_error(ytest, Ypredict))  
    print('Root Mean Squared Error:', np.sqrt(mean_squared_error(ytest, Ypredict)))
    print('R2 : ',r2_score(ytest, Ypredict))
    print()
```

```python
# Defining the keras model
model = Sequential()

# Adding the model layers
model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
```
